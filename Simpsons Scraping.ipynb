```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# get request
url = 'https://www.imdb.com/title/tt0096697/reviews?ref_=tt_ov_rt'
response = requests.get(url)
review_page = BeautifulSoup(response.content)

# get all review
review_part = review_page.findAll('div', attrs={'class':'lister-item-content'})
```

```python
# isolation per review

date_list = []
reviewer_name_list = []
rating_list = []
short_review_list = []
full_review_list = []
votes_list = []

for review in review_part:
    # review_date
    date = review.find('span', attrs={'class':'review-date'}).text
    print(f'review_date: :{date}')
    
    # reviewer's_name
    reviewer_name =  review.find('span', attrs={'class':'display-name-link'}).text
    print(f'reviewer_name: {reviewer_name}')
    
    # rating_score
    rating = review.find('span').text.strip()
    print(f'rating: {rating}')
    
    # short review
    short_review = review.a.text.strip()
    print(f'short review: {short_review}')
    
    # full review
    full_review = review.find('div', attrs={'class':'text show-more__control'}).text
    print(f'full_review: {full_review}')
    
    # number of votes of the review
    vote = review.find('div', attrs={'class':'actions text-muted'}).text.strip()
    votes = str(vote)[:13]
    print(f'number of votes: {votes}')
    print('\n')
    
    date_list.append(date)
    reviewer_name_list.append(reviewer_name)
    rating_list.append(rating)
    short_review_list.append(short_review)
    full_review_list.append(full_review)
    votes_list.append(votes)
    
    
```

```python
# build dataframe
df = pd.DataFrame({
    'date': date_list,
    'reviewer_name': reviewer_name_list,
    'rating': rating_list,
    'short_review': short_review_list,
    'full_review': full_review_list,
    'votes': votes_list
})
```
